```python
import numpy as np
import matplotlib.pylab as plt
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn import metrics
import sklearn
```

```python
df = pd.DataFrame(pd.read_excel('Austin_Housing_Data.xlsx'))

df.columns = ['Close Price', 'Green Building Verification', 'Green Energy Count', 'Green Energy Efficient', 
              'Green Sustainability', 'Accessibility Features', 'Bedroom Count', 'Garage Space Count', 
              'Homeowner Association', 'Latitude', 'Longitude', 'Occupant Type', 'Off Market Month', 
              'Off Market Year', 'Patio & Porch Features', 'Private Pool', 'Property Condition at Sale', 
              'Security Features', 'Square Feet', 'View', 'Year Built']


df
```

```python
df = df.dropna()


df = df.astype({"Close Price": int, "Green Building Verification": int, "Green Energy Count": int, 
                "Green Energy Efficient": int, "Green Sustainability": int, "Accessibility Features": int,
                "Bedroom Count": int, "Garage Space Count": int, "Homeowner Association": int, "Latitude": float, 
                "Longitude": float, "Occupant Type": int, "Off Market Month": int, "Off Market Year": int, 
                "Patio & Porch Features": int, "Private Pool": int, "Property Condition at Sale": int, 
                "Security Features": int, "Square Feet": int, "View": int, "Year Built": int})

# print(df.dtypes)

df['Close Price'] = np.log10(df['Close Price'])

df
```

```python
import xgboost as xgb
from sklearn.ensemble import RandomForestRegressor

from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error
```

```python
y = df['Close Price']

X = df[['Green Building Verification', 'Green Energy Count', 'Green Energy Efficient', 'Green Sustainability', 
      'Accessibility Features', 'Bedroom Count', 'Garage Space Count', 'Homeowner Association', 'Latitude', 
      'Longitude', 'Occupant Type', 'Off Market Month', 'Off Market Year', 'Patio & Porch Features', 'Private Pool', 
      'Property Condition at Sale', 'Security Features', 'Square Feet', 'View', 'Year Built']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 1)
```

```python
# define regressor
xg_regressor_test = xgb.XGBRegressor(eval_metric = 'rmse')

# set up search grid
parameter_grid = {"max_depth": [5, 6, 7, 8, 9, 10]}

# test every depth
search = GridSearchCV(xg_regressor_test, parameter_grid, cv=5).fit(X_train, y_train)

# print result
print("The best hyperparameter is", search.best_params_)
```

```python
# define regressor
xg_regressor = xgb.XGBRegressor(max_depth = 6)

# train regressor
xg_regressor.fit(X_train, y_train)

#generate predictions
predictions = xg_regressor.predict(X_test)

# find MSE
MSE1 = np.sqrt(mean_squared_error(y_test, predictions))
print("The mean squared error is %.5f" % MSE1 )
```

```python
# create data frame
df2 = pd.DataFrame()
df2['Features'] = df.columns[1:21]

# define regressor
xg = xgb.XGBRegressor(max_depth = 6)

# determine feature importances 
xg.fit(X,y)
xg_imp = xg.feature_importances_
df2['Score'] = xg_imp

# sort data frame
df2 = df2.sort_values(by = 'Score', ascending = False, ignore_index = True)

# display data frame
df2
```

```python
# define regressor
rf_regressor_test = RandomForestRegressor()

# set up search grid
parameter_grid = {"max_depth": [25, 26, 27, 28, 29, 30]}

# test every depth
search = GridSearchCV(rf_regressor_test, parameter_grid, cv=5).fit(X_train, y_train)

# print result
print("The best hyperparameter is", search.best_params_)
```

```python
# define regressor
rf_regressor = RandomForestRegressor(max_depth = 29)

# train regressor
rf_regressor.fit(X_train, y_train)

#generate predictions
predictions = rf_regressor.predict(X_test)

# find MSE
MSE2 = np.sqrt(mean_squared_error(y_test, predictions))
print("The mean squared error is %.5f" % MSE2 )
```
